```python
df_BRO= pd.read_table("fsnd_BRO_B.csv",   # 档名
                       encoding = 'utf8',       # 编码
                       header = 3,              # 指定数据的标题为原始数据之第3行
                       skiprows = [35,36],      # 原始数据之第35, 36行跳过不计
                       sep =','                 # 分隔符设为',' read_table内置分隔符为\t
                      )
df_hos= pd.read_table("fsnd_hos_A.csv",   # 档名
                       encoding = 'utf8',       # 编码
                       header = 3,              # 指定数据的标题为原始数据之第3行
                       skiprows = [35,36],      # 原始数据之第35, 36行跳过不计
                       sep =','                 # 分隔符设为',' read_table内置分隔符为\t
                      )
```

```python
df = df_hos
df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
years = [ x for x in df.columns if x!='地区'] 
df_melted = pd.melt(df, id_vars=['地区'], value_vars=years)
df_melted.columns = ['地区', '年', 'hos']
df_hos_done = df_melted
df_hos_done
```

```python
df = df_BRO
df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
years = [ x for x in df.columns if x!='地区'] 
df_melted = pd.melt(df, id_vars=['地区'], value_vars=years)
df_melted.columns = ['地区', '年', 'BRO']
df_BRO_done = df_melted
df_BRO_done
```

```python
df_p = df_BRO_done.join(df_hos_done,lsuffix='', rsuffix='_hos')
df_p
```

```python
df_p.drop(["地区_hos","年_hos"], axis=1)
```

```python
df_p.drop(["地区_hos","年_hos"], axis=1).set_index(["地区","年"])
```

```python
df_pp = df_p.drop(["地区_hos","年_hos"], axis=1).set_index(["地区","年"])
df_pp.query("年==2015")
```

```python
%matplotlib inline
df_pp.query("年==2015").plot(kind='scatter', x='hos', y='BRO', 
          title = 'hos与BRO的关系：分省数据')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False

df_plot = df_pp.query("年==2015")
df_plot.plot(kind='scatter', x='hos', y='BRO', 
          title = 'hos与BRO的关系：分省数据')
```

```python
import numpy as np
fit = np.polyfit(x=df_pp.query("年==2015")['hos'], y=df_pp.query("年==2015")['BRO'], deg=1)
```

```python
x=df_plot['hos']
y=df_plot['BRO']
ax = df_plot.plot(kind='scatter', x='hos', y='BRO', 
          title = 'hos与BRO的关系：分省数据')
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
```
